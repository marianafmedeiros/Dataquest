# Gender Gap in College Degrees

This project will focus on *visually* analyzing data of college degrees that can be found [here](https://github.com/fivethirtyeight/data/tree/master/college-majors).

As a first exploration of the dataset, we will plot the evolution of the graduate majors related to STEM through the years by Women and Men.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(20, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

Because there are 17 majors in total, we will generate a grid that has 6 rows and 3 columns. I will group the degrees into `stem_cats`, `liberal_arts_cats` and `other_cats`. Besides, the plots will be modified to have a better presentation and to induce better visual analysis.

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
liberal_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(20, 30))

for sp in range(1, 19, 3):
    i = 0 if sp == 1 else (int(sp/3))
    ax = fig.add_subplot(6,3,sp)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.set_yticks([0,50,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[i]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[i])
    ax.tick_params(bottom= False, top= False, left= False, right= False)
    if i == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif i == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(bottom= True)

for sp in range(2, 19, 3):
    i = 0 if sp == 2 else (int((sp)/3))
    if i <= 4:
        ax = fig.add_subplot(6,3,sp)
        ax.plot(women_degrees['Year'], women_degrees[liberal_arts_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[liberal_arts_cats[i]], c=cb_orange, label='Men', linewidth=3)
        ax.set_yticks([0,50,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        ax.set_title(liberal_arts_cats[i])    
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.tick_params(bottom= False, top= False, left= False, right= False)
        if i == 0:
            ax.text(2005, 80, 'Men')
            ax.text(2002, 20, 'Women')
        elif i == 4:
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            ax.tick_params(bottom= True)


for sp in range(3, 19, 3):
    i = 0 if sp == 3 else (int((sp)/3))
    if i <= 5:
        ax = fig.add_subplot(6,3,sp)
        ax.plot(women_degrees['Year'], women_degrees[other_cats[i]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[i]], c=cb_orange, label='Men', linewidth=3)
        ax.set_yticks([0,50,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        ax.spines["right"].set_visible(False)    
        ax.spines["left"].set_visible(False)
        ax.spines["top"].set_visible(False)    
        ax.spines["bottom"].set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(other_cats[i])
        ax.tick_params(bottom= False, top= False, left= False, right= False)
        if  i == 0:
            ax.text(2005, 90, 'Men')
            ax.text(2002, 8, 'Women')
        elif i == 5:
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            ax.tick_params(bottom= True)
        
plt.savefig('gender_degrees.png')
plt.show()
```

## Exporting the generated plots

We can export all of the plots contained in the figure as a single image file using `Figure.savefig()` or `pyplot.savefig()`. These have to be called before using `pyplot.show()` as this command destroys the internal state where the figure was being maintained. 
Above, the plots were saved as a single file called `'gender_degrees.png'`.
